# Konversi Data Series menjadi Supervised Learning

```python
# convert series ke supervised learning
from pandas import DataFrame
from pandas import concat
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]

    agg = concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
# memastikan semua data  float
values_Minyak = values_Minyak.astype('float32')
```

# Normalisasi Data menggunakan Min-Max

```python
# normalize features
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_Minyak = scaler.fit_transform(values_Minyak.reshape(-1,1))
```

```python
scaled_Minyak[:5]
```

```python
# frame as supervised learning
bingkai_Minyak = series_to_supervised(scaled_Minyak, 1, 1)
bingkai_Minyak
```

```python
# split into train and test sets
values = bingkai_Minyak.values
n_train = 292
train_Minyak = values[:n_train, :]
test_Minyak = values[n_train:, :]
```

```python
# split into input and outputs
train_X, train_y = train_Minyak[:, :-1], train_Minyak[:, -1]
test_X, test_y = test_Minyak[:, :-1], test_Minyak[:, -1]
```

```python
train_X.shape
```

```python
# reshape input to be 3D [samples, timesteps, features]
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from sklearn.model_selection import GridSearchCV
```

# Modelling LSTM

```python
# design model
model = Sequential()
model.add(LSTM(100, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# fit network
history = model.fit(train_X, train_y, epochs=100, batch_size=32, validation_data=(test_X, test_y), verbose=2, shuffle=False)
model.save('./savedModel')
```

```python
plt.plot(history.history['loss'],color='black', label='train')
plt.plot(history.history['val_loss'], color='gold', label='test')
plt.legend()
plt.show()
```

```python
result = model.evaluate(test_X, test_y)
result
```

```python
# make a prediction
yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))
```

```python
plt.plot(test_y, color= 'black',label="Data Aktual ")
plt.plot(yhat, color='gold', label="prediksi LSTM")
plt.legend(loc="best")
```

# Denormalisasi


```python
# invert scaling for predict
from numpy import concatenate
inv_yhat = concatenate((yhat, test_X[:, 1:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]
```

```python
# invert scaling for actual
test_y = test_y.reshape((len(test_y), 1))
inv_y = concatenate((test_y, test_X[:, 1:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]
```

#Forecasting Future Values

```python
import numpy as np
from tensorflow import keras

# Load your trained LSTM model
model = keras.models.load_model('./savedModel')

# Define the seed data (use the last few data points from your test dataset)
seed_data = test_y[-1:]  # Adjust 'look_back' according to your model

# Number of future time steps to forecast
num_forecast_steps = 7  # Adjust as needed

# Initialize a list to store the forecasted values
forecasted_values = []

# Generate forecasts iteratively
for _ in range(num_forecast_steps):
    # Reshape the seed data to match the input shape expected by the model
    seed_data_reshaped = np.reshape(seed_data, (-1, 1, 1))

    # Use the model to predict the next value
    next_value = model.predict(seed_data_reshaped)

    # Append the predicted value to the forecasted values list
    forecasted_values.append(next_value)

    # Update the seed data by removing the first element and adding the predicted value
    seed_data = np.append(seed_data[1:], next_value)

# The 'forecasted_values' list now contains your forecasts for future time steps
print(forecasted_values)
```
